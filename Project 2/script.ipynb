```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import prince
```

```python
# Task 1
df = pd.read_csv("data.csv")
print(df.shape, "\n")
print(df.info(), "\n")
print(df.describe(), "\n")
for col in df:
    print(col + ' ' + str(df[col].nunique()))
```

```python
# Task 2 - Data cleaning
# If you are starting from this task, you can run cells from all previous tasks in 
# the kernel by going to Kernel > Restart and Run All
print(df.iloc[0,6])
df.iloc[0,6] = df.iloc[0,6] + ' '
print(df.iloc[0,6])
```

```python
print(df.groupby(['Transmission Type']).size(), "\n") # frequency of each unique label in column
df['Transmission Type'].value_counts() # alternate and better way
```

```python
# remove leading and trailing whitespaces from each cell, only works on string columns
df = df.apply(lambda x: x.str.strip() if x.dtype=='object' else x)
df['Transmission Type'].value_counts()
```

```python
# check average NAs in each column
print(df.isna().mean())
df.isna().mean().plot(kind='barh')

# drop columns with >30% missing values (1 column). This cut-off is arbitrary, choose carefully
df = df.loc[:, df.isna().mean() < .3]
```

```python
# replace missing values with NaN, and drop rows with missing values
df = df.replace('NA',np.nan)
df = df.dropna()
df.info()
```

```python
print(df.duplicated().sum())
df = df.drop_duplicates()
df.info()
```

```python
# Task 3 - Patterns, outliers, correlations
# If you are starting from this task, you can run cells from all previous tasks in 
# the kernel by going to Kernel > Restart and Run All
fig, ax = plt.subplots(figsize=(12, 8))
sns.scatterplot(x="Engine HP", y="city mpg", data=df)
plt.xticks(rotation=45)
```

```python
# countplot
fig, ax = plt.subplots(figsize=(12, 8))
sns.countplot(x="Transmission Type", hue="Vehicle Style", data=df, ax=ax)
plt.xticks(rotation=45)
```

```python
#boxplot
fig, ax = plt.subplots(figsize=(12, 8))
sns.boxplot(x='Transmission Type', y='Engine HP', data=df)
plt.xticks(rotation=45)
```

```python
# pairplot - only numeric columns
numeric_cols = df.select_dtypes(include=np.number).columns.tolist()

#pairplot of first 5 numeric columns, coloured by Number of Doors
sns.pairplot(df, vars=numeric_cols[:5], palette='Set3')
```

```python
# Task 4 - find and visualize correlations
# If you are starting from this task, you can run cells from all previous tasks in 
# the kernel by going to Kernel > Restart and Run All
df_num = df[numeric_cols]
df_corr = df_num.corr()
print(df_corr, "\n")
print(df_corr.mean(), "\n")
print(df_corr.abs().mean(), "\n")
```

```python
# create heatmap of correlation
fig, ax = plt.subplots(figsize=(12, 8))
sns.heatmap(df_corr, square=True, vmin=-1, vmax=1, center=0,
            cmap=sns.diverging_palette(20, 220, n=200))
plt.xticks(rotation=45)
```

```python
# Task 5 - cluster your data to identify similar groups
# If you are starting from this task, you can run cells from all previous tasks in 
# the kernel by going to Kernel > Restart and Run All
print(df_num.columns.tolist())
X = StandardScaler().fit_transform(df_num)
kmeans = KMeans(n_clusters=5, init='random') # initialization
kmeans.fit(X) # actual execution
pred = kmeans.predict(X)
np.unique(pred)
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
plt.scatter(X[:, 1], X[:, 7], c=pred, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 1], centers[:, 7], c='grey', s=50)
```

```python
# Task 6 - PCA for dimensionality reduction
# If you are starting from this task, you can run cells from all previous tasks in 
# the kernel by going to Kernel > Restart and Run All
pca = PCA(n_components=0.95)
pca.fit(X)
pcad = pca.transform(X)
print(pca.explained_variance_ratio_)
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
sns.scatterplot(pcad[:,0], pcad[:,1])
```

```python
pca2 = prince.PCA(n_components=6, n_iter=3, rescale_with_mean=True,
    rescale_with_std=True, copy=True, engine='auto')
pca2 = pca2.fit(df_num)
pca2.explained_inertia_
```

```python
ax = pca2.plot_row_coordinates(df_num, ax=None, figsize=(12, 8),
     x_component=0, y_component=1, labels=None,
     color_labels=df['Transmission Type'], ellipse_outline=False,
     ellipse_fill=True, show_points=True
 )
```
